```python
import pandas as pd
import pandas_datareader as pdr
import datetime
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

start = datetime.datetime(2018,1,1)
end = datetime.datetime(2021,3,2)

Tesla = pdr.get_data_yahoo("TSLA", start, end)
GM = pdr.get_data_yahoo("GM", start, end)
Ford = pdr.get_data_yahoo("F", start, end)
Tesla['Close'].plot(title='Tesla VS GM VS Ford', figsize=(15,7)) # figsize in inch
GM['Close'].plot()
Ford['Close'].plot()
plt.legend()
plt.plot()
```

```python
Tesla['Volume'].plot(title='Tesla VS GM VS Ford', figsize=(15,7)) # figsize in inch
GM['Volume'].plot()
Ford['Volume'].plot()
plt.legend()
plt.plot()
```

```python
compare = pd.concat([Tesla['Open'], GM['Open'], Ford['Open']], axis=1) # axis=1 concate sideway, =0 by bottom
compare.columns = ['Tesla Open Price', 'GM Open Price', 'Ford Open Price']
scatter_matrix(compare, figsize=(15,7))
# print(compare)
```

```python
Tesla['Return'] = (Tesla['Close']-Tesla['Open'])/Tesla['Open'] * 100
GM['Return'] = (GM['Close']-GM['Open'])/GM['Open'] * 100
Ford['Return'] = (Ford['Close']-Ford['Open'])/Ford['Open'] * 100

Tesla['Return'].hist(figsize=(15,7),bins=30)
GM['Return'].hist(figsize=(15,7),bins=30,alpha=0.5)
Ford['Return'].hist(figsize=(15,7),bins=30,alpha=0.5)
```
